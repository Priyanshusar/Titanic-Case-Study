```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

titanic= sns.load_dataset("titanic")

titanic.head(5)
```

```python
# Create Histogram

plt.figure(figsize=(6,4))
sns.histplot(titanic['age'].dropna(),bins=10, kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

plt.figure(figsize=(6,4))
sns.histplot(titanic['fare'].dropna(),bins=10, kde=True)
plt.title("Fare Distribution")
plt.show()
```

```python
# Create Scatter plot

plt.figure(figsize=(6,4))
sns.scatterplot(x='age',y='fare',hue='survived',data=titanic)
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
# Create BOX plot

plt.figure(figsize=(6,4))
sns.boxplot(x='pclass',y='fare',data=titanic)
plt.title("Fare distribution by class")
plt.xlabel("PClass")
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(x='sex',y='age',data=titanic)
plt.title("Age distribution by sex")
plt.xlabel("Sex")
plt.show()
```

```python
# Create Heatmaps

plt.figure(figsize=(6,4))
sns.heatmap(titanic[['age','fare','pclass','sibsp']].corr(), annot=True ,cmap='coolwarm',linewidth=0.5)
plt.title("Heatmap of titanic Dataset")
plt.xlabel("Features")
plt.ylabel("Features")
plt.show()
```
